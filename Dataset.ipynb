```python
import numpy as np
import pandas as pd
df=pd.read_csv('SQLiV3.csv')
print(df)
```

```python
del df["Unnamed: 2"]
del df["Unnamed: 3"]
```

```python
print(df)
```

```python
d = pd.DataFrame(df)
print(d)
```

```python
d['LP']=d['Sentence'].map(str).apply(len)
d['NSPA']= d['Sentence'].str.count('\s+')
d['RSPA']=d['NSPA']/d['LP']
print(d)
```

```python
special = '[//*//"//+//−//∗/////=//!=//^=//<>//>=//<=//#///∗//∗///‘’//‘‘’’//@//\//()]'           
d['NSPE'] = d['Sentence'].str.count(special)
d['RSPE'] = d['NSPE']/d['LP']
```

```python
keyword = ["select","SELECT", "union", "UNION", "count", "COUNT", "group by", "GROUP BY", "order by", "ORDER BY", "insert", "INSERT", "delete", "DELETE", "update", "UPDATE", "drop table", "DROP TABLE", "truncate table", "TRUNCATE TABLE",
           "and", "AND", "or", "OR", "where", "WHERE", "from","FROM", "into", "INTO", "exec", "EXEC", "xp_cmdshell","XP_CMDSHELL", "master","MASTER", "net" , "NET","load_file", "LOAD_FILE", "outfile", "OUTFILE" ,"dumpfile", "DUMPFILE"]
d['NK'] = d['Sentence'].str.count(fr"\b(?:{'|'.join(keyword)})\b")
```

```python
weights = {"union":5, "UNION":5, "truncate":5, "TRUNCATE":5, "xp_cmdshell":5, "XP_CMDSHELL":5, "load_file":5, "LOAD_FILE":5, "outfile":5, "OUTFILE":5, "dumpfile":5, "DUMPFILE":5, "exec":5, "EXEC":5,
           "select":3, "SELECT":3, "update":3, "UPDATE":3, "insert":3, "INSERT":3, "delete":3, "DELETE":3, "count":3, "COUNT":3, "where":3, "WHERE":3, "group":3, "GROUP":3, "order":3, "ORDER":3, "drop":3, "DROP":3, "table":3, "TABLE":3, "master":3, "MASTER":3, "net":3 , "NET":3,
           "and":1,"AND":1, "or":1, "OR":1, "by":1, "BY":1, "from":1, "FROM":1, "into":1, "INTO":1}
d['KWS'] = d['Sentence'].str.split(expand=True).stack().map(weights).sum(level=0)
```

```python
d['ROC'] = 1-(d['RSPA']+d['RSPE'])
```

```python
d[d['Label'].apply(lambda x: str(x).isdigit())].to_csv('FinalDataset.csv', index= False)
```
